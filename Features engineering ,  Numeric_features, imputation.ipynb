## Random Sample imputation
#### Aim : RSI consist of taking random observation from the dataset , & we use these observation to replace Nan values
When it should be used : when data is MCAR


#### advantages>> variance of data is less
#### disadvantage >> In every situation it may not work


```python
df4 = pd.read_csv(r'D:\PYTHON Practice Projecs\train.csv', usecols=['Age','Fare','Survived'])
```

```python
df4.dtypes
```

```python
#% of null values
df4.isnull().mean()
```

```python
#### Replacing Nan values with some sampled values
df4['Age'].dropna().sample(df4['Age'].isnull().sum(),random_state= 0)
```

```python
def impute_nan(df,features,median):
    df[features+'_median'] = df[features].fillna(median)
    df[features+'_random'] = df[features]
    # Creating random sample
    random_sample = df[features].dropna().sample(df[features].isnull().sum(),random_state = 0)
    # Matching index
    random_sample.index = df[df[features].isnull()].index
    df.loc[df[features].isnull(),features+'_random']= random_sample

    
median = df4['Age'].median()    
impute_nan(df4,'Age',median)

## Plotting median & random imputation
fig = plt.figure()
ax = fig.add_subplot(111)
df4['Age'].plot(kind='kde', ax=ax)
df4.Age_median.plot(kind='kde', ax=ax, color='red')
df4.Age_random.plot(kind='kde', ax=ax, color='green')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

## End of distributon imputation




```python
df3 = pd.read_csv(r'D:\PYTHON Practice Projecs\train.csv', usecols=['Age','Fare','Survived'])
```

```python
# Third standard deviation
extreme = df3['Age'].mean()+3*df3['Age'].std()
```

```python
## There are outliers
sns.boxplot('Age',data=df3)
```

```python
def endof_dist(df,features,median,extreme):
    df3[features+'_endof_distribution'] = df3[features].fillna(extreme)
    df[features].fillna(median,inplace = True)    
```

```python
endof_dist(df3,'Age', df3['Age'].median(),extreme)
```
